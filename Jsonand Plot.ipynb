```python
import json

###################################################
def myPrintKeyValue(aKey, aValue):
    print("------")
    print("Key is " + str(aKey) + " Value is " + str(aValue) + " type of valule is " + str(type(aValue)))
###################################################
def myPrintDict(aDict):
    for dictKey, dictValue in aDict.items():
        myPrintKeyValue(dictKey, dictValue)
###################################################
def myPrintList(aList):
    for itemInList in aList:
        if (isinstance(itemInList, dict)):
            myPrintDict(itemInList)
        else:
            print("------")
            print(str(itemInList))
             
###################################################
# some JSON:
myJsonStr =  '{ "name":"John", "age":30, "city":"New York"}'

#I create xx as an multile level dictionary and will output it as json
myDict = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

# parse x:
xLoadedFromJson = json.loads(myJsonStr)

myDictOutputAsJson = json.dumps(myDict)
print("!!!Json looks like the follows!!!")
print(myDictOutputAsJson)
print(type(myDictOutputAsJson))
loadedBackFromMyDictJson = json.loads(myDictOutputAsJson)
print("loadedBackFromMyDictJson type is " + str(type(loadedBackFromMyDictJson)))
for myKey, myValue in loadedBackFromMyDictJson.items():
    if (isinstance(myValue, list)):
        myPrintList(myValue)
    else:
        myPrintKeyValue(myKey, myValue)
    



```

```python
# Converting JSON Data to a pretty format with ident or a Pandas DataFrame
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Step 1 : Request from the web page and saved the response from "get"
resp = requests.get('https://reqres.in/api/users')
# Step 2: Serialized the response using the .json() method to a dictionary
resp_dict = resp.json()

#Step 3 : You can do one or both the options below. 
#Option 1 Used the dumps() function with an indent of 4 to pretty print the response
#print("===Output pretty===")
pretty = json.dumps(resp_dict, indent=4)
print(pretty)
#Option 2, From step2, we passed the values of the 'data' key into the pd.DataFrame constructor, which accepts a list of dictionaries.
print("===Output the node, data, using panda data frame===")
df = pd.DataFrame(resp_dict.get('data'))
print(df)
#Step 4 delete the column without having to reassign df you can do:
df.drop(['email', 'avatar'], axis=1, inplace=True)
print("===Output after deleting column, avatar===")
print(df)
#Step 4, insert 2 new coumns, age as 2nd and income as 3rd column, savings as 4th column===")
print("===Output after adding 3 columns, age and income and savings===")
df.insert(1, "Age", [22, 33, 55, 36, 70, 52], True)
df.insert(2, "Income", [10000, 20000, 12000, 23000, 66000, 16800], True)
df.insert(3, "Savings", [10000, 10000, 30000, 36000, 58000, 20000], True)
print(df)
#Step 5, plotting
print("===Now plotting===")
df.plot(kind='area', x='Age', y=['Income', 'Savings'])
plt.show() #不叫plot.show因為我前面import plot as plt
```
